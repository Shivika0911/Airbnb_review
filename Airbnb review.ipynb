```python
import pandas as pd
df = pd.read_csv('/content/AirBNBReviews.csv')
```

```python
df.head(1)
```

```python
count = df['Positive or Negative'].value_counts()
```

```python
import matplotlib.pyplot as plt
plt.pie(count.values, labels=count.index, autopct = '%1.1f%%')
plt.title("Numbah of label")
plt.show
```

```python
snsdf = df.drop("Review", axis=1)
```

```python
snsdf['Genre'] = snsdf['Genre'].astype('category')
```

```python
import seaborn as sns
sns.catplot(x="Genre", hue="Positive or Negative", kind="count", data=snsdf)
```

```python
positive_df = snsdf[snsdf['Positive or Negative'] == 1]
sns.catplot(x="Genre", kind="count", data=positive_df)
plt.title('Positive Sentiment')
plt.show()

# Filter the DataFrame for negative sentiment (0)
negative_df = snsdf[snsdf['Positive or Negative'] == 0]

# Create a catplot for negative sentiment
sns.catplot(x="Genre", kind="count", data=negative_df)

# Show the plot for negative sentiment
plt.title('Negative Sentiment')
plt.show()
```
